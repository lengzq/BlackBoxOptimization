```python
%matplotlib inline
import AdamSPSA
import AdamRSGF
import numpy as np
import matplotlib.pyplot as plt
```

```python
# parameters for the algorithms
options = {}
options['a'] = 1e-1      
options['A'] = 0
options['a decay'] = 0.602
options['c'] = 1e-2
options['c decay'] = 0.101
options['gamma'] = 0.9
options['gamma decay'] = 0.5
options['beta'] = 0.3
options['beta decay'] = 0.0
options['delta'] = 1e-4
options['num repeat'] = 1
```

```python
def objective_function(theta):
    return theta**2
```

```python
adamspsa = AdamSPSA.AdamSpsa(options)
init_theta = np.array(1)
num_steps = 50
adamspsa.optimize(objective_function, init_theta, num_steps);
```

```python
result = adamspsa.get_optimization_params()
plt.figure(figsize=(5,5))
plt.plot(result['objective_values'], '-*')
plt.title('AdamSPSA')
plt.ylabel('Objective values')
plt.xlabel('Optimization steps')
```

```python
adamrsgf = AdamRSGF.AdamRsgf(options)
init_theta = np.array(1)
num_steps = 100
adamrsgf.optimize(objective_function, init_theta, num_steps);
```

```python
result = adamrsgf.get_optimization_params()
plt.figure(figsize=(5,5))
plt.plot(result['objective_values'], '-*')
plt.title('AdamRSGF')
plt.ylabel('Objective values')
plt.xlabel('Optimization steps')
```
